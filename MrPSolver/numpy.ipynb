## Creare array
Ci sono diversi modi per creare un array in NumPy. Il più utilizzato è il metodo `np.array()`

```python
arr1 = np.array([1, 2, 3, 4])
arr1
```

Per creare un array riempito con zeri o uno, si usano rispettivamente

```python
N_elements = 10
arr2 = np.zeros(N_elements)
arr2
```

```python
arr3 = np.ones(N_elements)
arr3
```

Si possono anche creare array di numeri pseudo-casuali, usando `np.random.*()`

```python
arr4 = np.random.random(N_elements) # Array di numeri distribuiti uniformemente tra 0 e 1
arr4
```

```python
arr5 = np.random.randn(N_elements) # Array di numeri gaussiani con mu = 0 e sigma = 1
arr5
```

E' anche possibile creare array di elementi con una certa spaziatura costante; questi sono particolarmente utili per plottare funzioni con MatPlotLib. Con `np.linspace()` si definisce il range degli elementi da rappresentare e il numero degli elementi da rappresentare, e viene restituito un array contenente i punti da `x_min` a `x_max` automaticamente spaziati.

```python
x_min = 0.0
x_max = 100.0
arr6 = np.linspace(x_min, x_max, N_elements)
arr6
```

Se invece si è interessati alla spaziatura tra gli elementi, si usa `np.arange()`

```python
spacing = 0.5
arr7 = np.arange(x_min, x_max, spacing)
arr7
```

## Operazioni sugli array
In NumPy è possibile svolgere diverse operazioni sugli array, o tra array. Di seguito sono riportati alcuni esempi, la lista completa si può trovare al seguente [link](https://numpy.org/doc/stable/reference/routines.html)

```python
# Moltiplica tutti gli elementi di un array per un fattore
2 * arr1
```

```python
# Reciproco di ciascun elemento
1 / arr1
```

```python
# Somma un valore a tutti gli elementi
arr1 + 6.5
```

```python
# Calcola il quadrato di ciascun elemento e plotta il grafico
y = arr7**2
plt.plot(arr7, y)
plt.show()
```

Si possono anche comporre funzioni, in questo modo si può evitare di usare cicli for

```python
def f(x):
    return x**2 * np.sin(x) / np.exp(-x)

x = np.linspace(0, 10, 100)
y = f(x) # f viene applicata a ciascun elemento dell'array

plt.plot(x, y)
plt.show()
```

## Indexing e slicing
Per estrarre un elemento dall'array

```python
arr = np.array([2, 4, 6, 8, 10])
arr[2]
```

Per fare uno slicing

```python
arr[2:] # Prende solo gli elementi a partire dall'indice 2 (compreso)
```

```python
arr[:-2] # Prende solo gli elementi fino al terzultimo elemento
```

```python
arr[1:-2] # Combinazione dei due precedenti
```

E' molto utile filtrare gli elementi di un array in base ad una proprietà, per esempio

```python
arr > 3 # Restituisce True solo se l'elemento dell'array è maggiore di 3, altrimenti restituisce False
```

A questo punto si può usare l'array di valori booleani ottenuto per indicizzare l'array di partenza, ottenendo un nuovo array che contiene soltanto gli elementi che soddisfano la proprietà richiesta

```python
arr[arr > 3]
```

Questo è particolarmente utile quando si maneggiano array con molti dati. Per esempio, dato un array di nomi si vuole ottenere un array contenente soltanto i nomi che iniziano per "J". Usando la funzione `np.vectorize()`, si può applicare una qualsiasi funzione a ciascun elemento dell'array

```python
names = np.array(["Jim", "Jake", "Luke", "Matthew", "Josh", "Peter"])
first_letter = np.vectorize(lambda s: s[0])(names)
first_letter
```

In questo modo, l'array `first_letter` conterrà solo la prima lettera di ciascun nome. Ora si deve controllare se la prima lettera è "J"

```python
first_letter_j = (first_letter == 'J')
first_letter_j
```

Infine si indicizza l'array

```python
names_j = names[first_letter_j]
names_j
```

Un altro esempio è trovare gli elementi pari di un array

```python
numbers = np.array([357, 302, 11, 28, 12, 19, 74, 47, 234])
is_even = (numbers % 2 == 0)
numbers[is_even]
```

## Applicazioni alla statistica e al calcolo

### Statistica
Per esempio, si possono calcolare media, deviazione standard e percentile di un sample

```python
sample = 1.5 * np.random.randn(10000) + 15.0 # Numeri gaussiani con mu = 15 e sigma = 1.5
plt.hist(sample, color = "green")
plt.show()
```

```python
mu = np.mean(sample) # Media
sigma = np.std(sample) # Deviazione standard
print(f"Media: {mu}, Std: {sigma}")
```

```python
perc = np.percentile(sample, 80) # 80% percentile
perc
```

### Integrali e derivate
NumPy può essere utilizzato per il calcolo numerico di derivate o integrali

```python
x = np.linspace(1, 10, 100)
y = 1 / x**2 * np.sin(x) # Funzione di partenza
dy_dx = np.gradient(y, x) # Derivata rispetto ad x
dx = x[1] - x[0]
integral_y = np.cumsum(y) * dx # Integrale rispetto ad x usando somma cumulativa

plt.plot(x, y)
plt.plot(x, dy_dx)
plt.plot(x, integral_y)
plt.show()
```

### Esercizio 3
Si considera la curva $ r(\theta) = 1 + \frac{3}{4} \sin(3 \theta) $ per $ \theta \in [0, 2 \pi] $.

1. Plotta la curva nel piano
2. Calcola l'area della regione delimitata dalla curva, usando la formula

$$ A = \int_{0}^{2 \pi} \frac{1}{2} r^2 d \theta $$

3. Calcola la lunghezza della curva, ricordando che

$$ L = \int_{0}^{2 \pi} \sqrt{r^2 + \left( \frac{dr}{d \theta} \right)^2 } d \theta $$

```python
def r(theta):
    return 1 + (3 / 4) * np.sin(3 * theta)
```

```python
# Punto 1
theta_vals = np.linspace(0, 2 * np.pi, 1000)
r_vals = r(theta_vals)

x_vals = r_vals * np.cos(theta_vals)
y_vals = r_vals * np.sin(theta_vals)

plt.plot(x_vals, y_vals, color = "red")
plt.show()
```

```python
# Punto 2
d_theta = theta_vals[1] - theta_vals[0]
integral = np.cumsum(0.5 * r_vals**2) * d_theta
integral[-1]
```

```python
# Punto 3
dr_dtheta = np.gradient(r_vals, theta_vals)
integral = np.cumsum(np.sqrt(r_vals**2 + dr_dtheta**2)) * d_theta
integral[-1]
```

### Esercizio 4
La potenza irradiata da un corpo nero è data da $ P = A \sigma \epsilon T^4 $ (legge di Stefan-Boltzmann). Dopo aver misurato la temperatura di una stella, si trova che

$$ T(t) = \frac{T_0}{1 + e^{-kt}} $$

Plotta l'energia totale emessa dalla stella come funzione di $ t $ usando

$$ E(t) = \int_{t' = 0}^{t' = t} P(t') dt' $$

#### Svolgimento
Si ha che

$$ P(t) = \frac{A \sigma \epsilon T_0^4}{(1 + e^{-kt})^4} $$

Il problema è che le quantità $ A $, $ \sigma $, $ \epsilon $ e $ T_0 $ non sono note. Per aggirarlo, si lavora con la quantità adimensionata 

$$ \frac{P(t)}{A \sigma \epsilon T_0^4} = \frac{1}{(1 + e^{-kt})^4} $$

Inoltre, visto che non si conosce $ k $, si usa $ kt $ invece di $ t $ come variabile indipendente (anch'essa adimensionata).

```python
kt = np.linspace(0, 3, 100)
P = 1 / (1 + np.exp(-kt))**4
d_kt = kt[1] - kt[0]
E = np.cumsum(P) * d_kt

plt.plot(kt, P)
plt.plot(kt, E)
plt.xlabel("$kt$", fontsize = 20)
plt.ylabel(r"$\left(\frac{k}{A \sigma \epsilon T_0^4} \right) E(kt)$", fontsize = 20)
plt.show()
```

## Array multi-dimensionali
Per creare un array multi-dimensionale

```python
arr1 = np.array([[4, 6, 4], [1, 2, 1], [3, 1, 1]])
arr1
```

Le operazioni viste per il caso unidimensionale funzionano allo stesso modo, per esempio

```python
2 / arr1
```

Anche l'indexing booleano funziona allo stesso modo

```python
arr1 > 2
```

```python
arr1[arr1 > 2]
```

In due dimensioni, la sintassi per lo slicing è un poà più complicata. Per estrarre una riga

```python
arr1[1] # Seconda riga
```

Per estrarre una colonna

```python
arr1[:,1] # Seconda colonna
```

":" significa sostanzialmente "considera ogni riga della matrice". Poi, separato da una virgola, "1" dice di prendere il secondo elemento di ciascuna riga. Il risultato ottenuto è la seconda colonna.

Per prendere gli ultimi due elementi della seconda colonna

```python
arr1[1:,1]
```

Per prendere la sottomatrice 2x2 in basso a sinistra

```python
arr1[1:, :2]
```

Infine, il metodo `np.ravel()`, molto utile, viene utilizzato per collassare un array multi-dimensionale in un array monodimensionale

```python
np.ravel(arr1)
```
